```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# import data
df = pd.read_csv("C:/Users/Utku/Desktop/Jupyter Notebook/Datathon/Datasets/linear_regression_dataset_20220209.csv")
df.head()
```

```python
# data visualization
plt.scatter(df.experience, df.salary)
plt.xlabel("Experience (Year)")
plt.ylabel("Salary ($)")
plt.title("Relation of Experience - Salary")
plt.grid(True)
plt.show()
```

```python
# Linear Regression Model Training

# Linear Regression Model
linear_reg = LinearRegression()

# changing experience and salary data to numpy array
x = df.experience.values.reshape(-1, 1)
y = df.salary.values.reshape(-1, 1)

# Linear Regression Training
linear_reg.fit(x, y)
```

```python
# the point where it intersects the y-axis (alternate)
y_axis_intersect = np.array([0]).reshape(1, -1)
b0 = linear_reg.predict(y_axis_intersect)
print("b0: ", b0)

# the point where it intersects the y-axis
b0_ = linear_reg.intercept_
print("b0: ", b0_)

# slope
b1 = linear_reg.coef_
print("b1: ", b1)
```

```python
# salary = 1663 + 1138 * experience

experience = 11

newSalary = 1663 + 1138 * experience
print(newSalary)

# with predict method
result = linear_reg.predict(np.array([experience]).reshape(1, -1))
print(f"Salary of someone with 11 years of experience: {result[0]}")
```

```python
# Test / Prediction / Visualization with Linear Regression Model
array = np.array([range(0, 16)]).reshape(-1, 1)

plt.figure()
plt.scatter(x, y)

# Predicting salary of someone with 0-15 years of experience
y_head = linear_reg.predict(array) # y_head = salary

plt.plot(array, y_head, color= "r") # (experience, salary)
plt.xlabel("Experience (Year)")
plt.ylabel("Salary ($)")
plt.title("Relation of Experience - Salary")
plt.grid(True)
plt.show()
```
