```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

# histogram of the datas
n, bins, patches = plt.hist(x, 100, density = 1, facecolor= 'b', alpha= 0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor= 'g', alpha= 0.25)

plt.xlabel('Datas')
plt.ylabel('Possibilities')
plt.title(r'Histogram of the Datas: $\mu_1=100,\ \mu_2=110$')
plt.annotate('Top of the Gaussion Curve', xy= (110, 0.04), xytext= (130, 0.045),
             arrowprops= dict(facecolor= 'black', shrink= 0.1)) # creates an arrow
plt.text(75, 0.025, r'$\mu_1=100,\ \sigma_1=15$')
plt.text(120, 0.035, r'$\mu_2=110,\ \sigma_2=10$')
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
```

```python
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected= True)
import plotly.graph_objs as go
import plotly.io as io
import plotly.figure_factory as ff
```

```python
data = pd.read_csv("C:/Users/Utku/Desktop/Jupyter Notebook/Datathon/04.02.2022 Data Visulization with Python/University_Rankings_20220204.csv")
data.head(3)
```

```python
data.columns
```

## Line Graph

```python
df = data.iloc[:100, :]
df.head(3)
```

```python
# plotly styles
io.templates.default = 'simple_white'
```

```python
line1 = go.Scatter(x = df.world_rank,
                   y = df.citations,
                   mode = 'markers',
                   name = 'Citations',
                   marker = dict(color = 'rgba(78, 78, 250, .85)'),
                   text = df.university_name)

line2 = go.Scatter(x = df.world_rank,
                   y = df.teaching,
                   mode = 'markers',
                   name = 'Teaching',
                   marker = dict(color = 'rgba(202, 43, 15, .85)'),
                   text = df.university_name)

data = [line1, line2]
location = dict(title = "Citation and Teaching Points of World Top 100 University",
                xaxis = dict(title='World Ranking', ticklen=5, zeroline=False))

fig = dict(data=data, layout=location)
plot(fig, filename = '1_Line_Citation_Teaching_Points.html')
```

## Bar Graph

```python
data2011 = df[df.year == 2011].iloc[:5,:]

bar1 = go.Bar(x = data2011.university_name,
                y = data2011.citations,
                name = 'Citations',
                marker = dict(color = 'rgba(255, 127, 40, .5)',
                line = dict(color = 'rgb(0, 0, 0)', width = 1.5)),
                text = data2011.country)

bar2 = go.Bar(x = data2011.university_name,
                y = data2011.teaching,
                name = 'Teaching',
                marker = dict(color = 'rgba(64, 10, 128, .5)',
                line = dict(color = 'rgb(0, 0, 0)', width = 1.5)),
                text = data2011.country)

data_ = [bar1, bar2]
location = go.Layout(barmode = 'group')
fig = go.Figure(data=data_, layout = location)
plot(fig, filename = '2_Bar_Citation_Teaching_Points.html')
```

## Pie

```python
data2011 = df[df.year == 2011].iloc[:11,:]

slice1 = data2011.num_students
slice1_list = [float(each.replace(',','.')) for each in data2011.num_students]
tags = data2011.university_name

pie = go.Pie(labels= tags,
               values = slice1_list,
               hoverinfo = 'label+value+percent',
               textinfo = 'value+percent',
               textfont =dict(size =8),
               rotation = 100,
               hole= 0.3,
               marker = dict(line=dict(color='#000000', width =1)))

data_ = [pie]
location = dict(title= 'Number of the Students of Top 8 Universities in the World in 2011',
                legend=dict(orientation='h'))
fig = dict(data=data_, layout= location)
plot(fig, filename = '3_Pie_NumberofStudents.html')
```

## Bubble Graph

```python
data2011 = df[df.year == 2011].iloc[:20,:]

num_students = [float(each.replace(',','.')) for each in data2011.num_students]
international_color =[float(each) for each in data2011.international]

data = [{'y': data2011.teaching,
         'x': data2011.world_rank,
         'mode': 'markers',
         'marker': {'color': international_color,
                    'size': num_students,
                    'showscale': True},
         'text': data2011.university_name}]

plot(data, filename= '4_Bubble_NumberofStudents.html')
```
